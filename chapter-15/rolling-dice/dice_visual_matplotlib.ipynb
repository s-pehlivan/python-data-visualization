```python
import matplotlib.pyplot as plt

from die import Die

# Create 3 D6
die_1 = Die()
die_2 = Die()
die_3 = Die()

# Roll the dice and save the results and their frequencies
results = [die_1.roll() + die_2.roll() + die_3.roll() for val in range(50_000)]
max_value = die_1.num_sides + die_2.num_sides + die_3.num_sides

# Visualize data
plt.figure(figsize=(7,7))

fig, ax = plt.subplots()

plt.hist(results, 
        bins=range(3, max_value + 1),
        color="skyblue",
        edgecolor=(0, 0, 0, 0.5),
        linewidth=0.2)

ax.set_title("The frequencies of results when a three D6 is rolled")
ax.set_xlabel("Results")
ax.set_ylabel("Frequencies of Results")

plt.show()


```
