```python
import plotly.graph_objects as go
import numpy as np
from random_walk import RandomWalk

rw = RandomWalk(50_000)
rw.fill_walk()

# color gradient with indices
num_points = len(rw.x_values)
colors = np.linspace(0,1,num_points)

color_scale = [[0, "lightblue"], [1, "darkblue"]]
marker_colors = colors

fig = go.Figure(data=go.Scatter(
    x=rw.x_values, 
    y=rw.y_values,
    mode="markers",
    marker= dict(
        size = 3,
        color = marker_colors,
        colorscale = color_scale,
        showscale = False
        
    )
))

fig.add_trace(go.Scatter(
    x=[rw.x_values[0]],
    y=[rw.y_values[0]],
    mode="markers",
    marker=dict(
        size = 10,
        color = "green"
    ),
    name = "Start"
))

fig.add_trace(go.Scatter(
    x=[rw.x_values[-1]],
    y=[rw.y_values[-1]],
    mode = "markers",
    marker = dict(
        size= 10 ,
        color = "red"
    ),
    name = "End"
    
))

fig.update_layout(
    title="Random Walk Pattern", 
    width = 991,
    height = 991,
    showlegend = True)

fig.show()


```
